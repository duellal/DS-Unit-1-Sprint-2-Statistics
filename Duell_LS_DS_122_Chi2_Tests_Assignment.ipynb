- What is a sample "estimate" in statistics?

A sample estimate is a small size of the population that you are looking at for a variable. It should give you a good estimate of the variable for the whole population.

- What are hypothesis test? How are they useful?

Hypothesis tests are tests where you see if the results of a certain study have happened by chance, aka: the validity of the claim. This is testing the null hypothesis and seeing if there is significant data to back up that the claim is true or not. There are 5 steps: 1. Creating a Null hypothesis, 2. Creating an alternative hypothesis, 3. Stating the confidence level, 4. Calculating the t-statistic and the p-value, and 5. Stating the conclusion.

- What is a "null hypothesis?"

A null hypothesis states that there is no relationship in the given data values. This reflects that any relationship in the sample is based on chance.

- What is a p-value and what does it represent?

The p-value is the probability for the null hypothesis. The larger the p-value, you should accept the null hypothesis. The lower the p-value, you should reject the null hypothesis and choose the alternative hypothesis. You compare your p-value to the confidence level that you selected in step 3.

- What does it mean for something to be "statistically significant?"

For something to be "statistically significant" it means that, in the case of the p-value, the probability that the null hypothesis is true is very slim when the p-value is small compared to the confidence level.

- What is a test statistic and how does it relate to a p-value?

A test statistic takes the data from the sample and compares it to what was expected. The lower the test statistic, the more likely you will accept the null hypothesis. The higher the test statistic, the more likely you will reject the null hypothesis and accept the alternative hypothesis.

- What are "degrees of freedom" and how are they calculated in a 1-sample, 2-sample, and chi^2 test for independence?

Degrees of freedom is the sample size you are using. In a 1 sample test for independence, the degrees of freedom are calculated by subtracting 1 from the sample size; $n - 1$. In a 2 sample test for independence, the degrees of freedom are calculated by adding the two sample sizes together and subtracting by 2; $(n_1 + n_2) - 2$.
In a $\chi^2$ sample test for independence, the degrees of freedom are calculated by the number of rows minus 1, times the number of columns minus 1; $r = rows, c = columns: (r-1)(c-1)$.
